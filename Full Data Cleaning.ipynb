```python
housecount = finaldf.groupby("day")[["LCLid"]].count()
housecount.rename(columns={"LCLid":"housecount"}, inplace=True)
housecount.head()
```

```python
finaldfEnergySum=finaldf.groupby(["day"])[['TotEnergyPerDay']].sum()
```

```python
finaldfmean = finaldf.groupby("day").mean()
finaldfmean.drop(["TotEnergyPerDay"], axis=1, inplace = True)
finaldfmean.holiday = finaldfmean.holiday.astype(int)
```

```python
finaldfdata=finaldfmean.merge(housecount,on="day")
finaldfdata=finaldfdata.merge(finaldfEnergySum,on="day")
finaldfdata.head()
```

```python
finaldfdata["AvgEnergyPerDay"] = [i / j for i, j in zip( finaldfdata["TotEnergyPerDay"], finaldfdata["housecount"])]
finaldfdata.head()
finaldfdata.tail()
```

```python
finaldfdata.reset_index(inplace = True)
```

```python
finaldfdata = finaldfdata[finaldfdata["day"]!="2014-02-28"]
finaldfdata.tail()
```

```python
finaldfdata['day'] = pd.to_datetime(finaldfdata["day"])
finaldfdata["month"] = finaldfdata["day"].dt.month
finaldfdata["date"] = finaldfdata["day"].dt.day
```
